```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
import seaborn as sns
```

```python
# task 
days=["Mon","Tue","Wed","Thus","Fri"]
cities=["New York","Londan","Delhi","Tokyo"]
temp=[
    [22,23,21,24,25],
    [21,18,19,23,26],
    [30,35,36,32,31],
    [25,26,23,25,26]
]
```

## - Task 
### -Lineplot of days v/s temp for each city in a single figure

### METHOD 1 

```python
plt.style.use("dark_background")
fig ,ax =plt.subplots(2,2,figsize=(7,7))
city_no=0
for i in range(2):  # 0 , 1
    for j in range(2):  # 0 ,1 
        ax[i][j].plot(days,temp[city_no])
        ax[i][j].set_xlabel("Days")
        ax[i][j].set_ylabel("Temperature")   
        ax[i][j].set_title(f'city {cities[city_no]}')
        ax[i][j].grid(True)
        city_no+=1

fig.tight_layout()
```

## METHOD 2

```python
# flatting the axes 
figure , axes = plt.subplots(2,2 ,figsize=(8,8))

axes = axes.flatten()
for idx , ax in enumerate(axes):
    ax.plot(days,temp[idx],"go-")
    ax.set_title(cities[idx])
    ax.set_xlabel("days")
    ax.set_ylabel("temperature")
    ax.grid(True)
figure.tight_layout()
figure.suptitle("Temperature in Cities Over the Week")
figure.supxlabel("Days")
figure.supylabel("Temprature")
```
