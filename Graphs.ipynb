```python
from sklearn.datasets import make_regression
```

```python
x,y= make_regression(
    n_features=5,
    n_samples=10,
    noise=0.2
)
```

```python
# graph 
# note x was 5 dimenstions when also its is clear

plt.figure(figsize=(5,5))
plt.plot(x[:,0],y ,"bo-" ,label="x1")
plt.plot(x[:,1],y ,"g>-",label="x2")
plt.plot(x[:,2],y ,"r<-", label="x3")
plt.plot(x[:,3],y ,"w--", label="x4")
plt.grid(True)
plt.style.use('dark_background')
plt.xlabel("features values")
plt.ylabel("target values")
plt.title("regression graph")
plt.legend()
```

## Box plot

### box plot is used to visualize 5 main in data
- Minimun (lowest non-outlier)
- first quartile(Q1) - 25th percentile
- Median(Q2) - 50th percentile
- Third Quartile(Q3) - 75th percentile
- Maximun (Highest non-outlier)

#### find the Whiskers(max non outlier) and Outlier value

```python
# used to detect some outliears
sns.boxplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex",
    showmeans=True,
    whis=2.0
)
plt.style.use("classic")
plt.ylim(0,9)

# UpperWhisker --> Q3 + (1.5*IQR)  IQR = box distance (Q3-Q1) -->  nearst datapt.in dataset
# lowerWhisker --> Q1 - (1.5*IQR)  IQR = box distance (Q3 - Q1)
```

# HeatMap

```python
# these are the matrix plots 
# plot rectangular data as color-encoded matrixs - 2d metrcis
```

```python
flights
```

```python
# transform flights dataframe with pivot(concat widely -rectangular form ) and melt (long format)method    
flight_pivot=flights.pivot(index="month",columns="year" ,values="passengers")
```

```python
# problem of extracting passengers values min , max
sns.heatmap(data=flight_pivot,
           cmap="crest",
           annot=True,
           robust=False,
           fmt="d",
           center=True,
           linecolor="black",
           linewidths=0.2)# format 

# cmap : crest , coolwarm , default
# annot: True 
# fmt : (d) string , (.1f) ,(.2g) for decimal values 
plt.title("HeatMap")
plt.tight_layout()
```
